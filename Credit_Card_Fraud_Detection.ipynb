```python
# Importing necessary packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading data
data = pd.read_csv('creditcard.csv')
data.head()
```

```python
# Exploring the dataset
print(data.columns)
print(data.shape)
```

```python
print(data.describe())
```

```python
# Sampling the dataset 
data = data.sample(frac = 0.1, random_state=1)
print(data.shape)
```

```python
# Plotting histogram for each column
data.hist(figsize = (14,14))
plt.show()
```

```python
# Determining number of fraud cases in dataset

fraud_cases = data[data['Class'] == 1]
valid_cases = data[data['Class'] == 0]
```

```python
outlier_fraction = len(fraud_cases) / len(data['Class']) * 100
```

```python
print('Fraud Cases : {}'.format(len(fraud_cases)))
print('Valid Cases : {}'.format(len(valid_cases)))
outlier_fraction
```

```python
# Correlation matrix
corrmat = data.corr()
fig = plt.figure(figsize = (20,15))
sns.heatmap(corrmat, annot = True, vmax = .8, square=False)
plt.show()
```
